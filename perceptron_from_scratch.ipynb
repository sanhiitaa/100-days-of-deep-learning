```python
# creating syntetic data

from sklearn.datasets import make_classification
import numpy as np

X,y= make_classification(n_features=2,
                         n_samples=100,
                         n_informative=1,
                         n_redundant=0,
                         n_classes=2,
                         n_clusters_per_class=1,
                         random_state=999,
                         hypercube=False,
                         class_sep=10)
```

```python
# visualizing the data generated

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='Set1', s=100)
```

```python
# defining step function
def step (z):
  return 1 if z>0 else 0
```

```python
# defining perceptron

def perceptron(X,y):
  X=np.insert(X,0,1,axis=1)
  weights=np.ones(X.shape[1])
  lr=0.1

  for i in range(1000):
    j= np.random.randint(0,100)
    y_pred= step(np.dot(X[j],weights))
    weights= weights + lr*(y[j]-y_pred)*X[j]

  return weights[0], weights[1:]

```

```python
intercept, coefficient= perceptron(X,y)
```

```python
print('intercept: ', intercept)
print('coefficient: ', coefficient)
```

```python
m=-(coefficient[0]/coefficient[1])
b=-(intercept/coefficient[1])
```

```python
# creating a line from the values provided by the perceptron
x_input= np.linspace(-3,3,100)
y_input=m*x_input + b
```

```python
# plotting the line
plt.figure(figsize=(10,6))
plt.plot(x_input, y_input, color='red', linewidth=3)
plt.scatter(X[:,0], X[:,1], c=y, cmap='Set1', s=100)
plt.ylim(-3,2);
```
