```python
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
!mkdir train/ # stores the images that are to be augmented
!mkdir aug/   # stores post augmentation images
!mv '/content/redpanda.jpg' train/ # moving the sample image to the train folder
```

```python
img = image.load_img('/content/train/redpanda.jpg', target_size=(200,200))
```

```python
print(type(img))
img
```

```python
# converting img into an array
img= image.img_to_array(img)
img.shape
```

```python
# reshaping the image to create a batch with a single image
input_batch=img.reshape(1,200,200,3)
```

```python
# creating an object of the ImageDataGenerator class
datagen= ImageDataGenerator(rotation_range=50,
                            width_shift_range=0.5,
                            height_shift_range=0.3,
                            shear_range=0.3,
                            zoom_range=0.4)
```

As we are working with a single image in this case, we will use the `flow` method of the `ImageDataGenerator` class, if we are to pass a whole directory of images we will use `flow_from_directory` method.

```python
i=0

for output in datagen.flow(input_batch,batch_size=1, save_to_dir='aug'):
  #necessary to define a break condition as this is an infinite for loop
  i+=1
  if i==10:
    break
```
