```python
# creating synthetic data
from sklearn.datasets import make_classification
import numpy as np

X,y= make_classification(n_samples=100,
                         n_features=2,
                         n_informative=1,
                         n_redundant=0,
                         n_classes=2,
                         n_clusters_per_class=1,
                         random_state=41,
                         hypercube=False,
                         class_sep=15
                         )
```

```python
# visualizing the data generated

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='Set1', s=50);
```

```python
# defining perceptron with hinge loss

def perceptron_w_hinge(X,y):
  w1=w2=b=1
  lr=0.1

  for i in range(2000):
    for j in range(X.shape[0]):

      z=w1*X[j][0] + w2*X[j][1] + b

      #checking condition
      if z*y[j] <0:
        w1 = w1 + lr * y[j] * X[j][0]
        w2 = w2 + lr * y[j] * X[j][1]
        b = b + lr * y[j]
  return w1, w2, b

```

```python
w1, w2, b = perceptron_w_hinge(X,y)
# w1, w2, b = perceptron(X,y)
```

```python
w1, w2, b
```

```python
m=-(w1/w2)
c=-(b/w2)

m,c
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c

plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='Set1',s=50)
plt.ylim(-3,2)
```
