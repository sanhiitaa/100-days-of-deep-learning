## Importing Libraries

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import numpy as np
```

## Tokenizing our text

```python
token= Tokenizer()
```

```python
token.fit_on_texts([text, text1])
```

```python
token.word_index
```

```python
len(token.word_index)
```

```python
for sentence in text.split('\n'):
  print(sentence)
```

```python

input_sequences=[]

for doc in [text, text1]:
  for sentence in doc.split('\n'):
    tokenized_sentence = token.texts_to_sequences([sentence])[0] # converting sentences to their respective sequences
    print(tokenized_sentence) # for visualization purpose

    #
    for i in range(1, len(tokenized_sentence)):
      n_gram_sequence= tokenized_sentence[:i+1]
      input_sequences.append(n_gram_sequence)
```

```python
print('total input sequences after creating n_gram sequences: ', len(input_sequences), '\n\n**********************************')
print('n_gram sequences: \n')
input_sequences
```

## Zero Padding the data

```python
# finding the length of the longest sequence
maxlen=max([len(x) for x in input_sequences])
```

```python
# padding the seuqences based on the longest sentence
padded_sequences = pad_sequences(input_sequences, maxlen=maxlen, padding='pre')
```

```python
padded_sequences
```

```python
len(padded_sequences)
```

We will use supervised learning to perfrom the next word prediction, classification task specifically.

```python
# splitting the data into dependent and independent variables
x= padded_sequences[:, :-1]
y= padded_sequences[:, -1]
```

```python
# shapes of the variables
x.shape, y.shape
```

1. `x.shape = (num_samples, maxlen-1)`
2. `y.shape = (num_samples, num_classes).`

```python
num_classes= len(token.word_index) + 1 # +1 because indexing starts from one and OHE starts from 0

# OHE the categorical variables.
y = to_categorical(y, num_classes=num_classes)
```

```python
# updated shape of the dependent variable
y.shape
```

# Building LSTM Architecture

For reference:
1. `num_classes` : total number of unique words + 1
2. `maxlen` : length of the longest sentence

```python
# building the model

model = Sequential()
model.add(Embedding(num_classes, 100, input_length=maxlen-1)) # -1 because we exclude the last token, it is the first output of y
model.add(LSTM(200))
model.add(Dense(num_classes, activation='softmax'))
```

```python
model.summary()
```

```python
# compiling the model
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# fitting the model
model.fit(x,y, epochs=100)
```

## Prediction

```python
# text from training data
text='Believe'

for i in range(7):
  # tokenize
  token_text = token.texts_to_sequences([text])[0]

  # padding
  padded_token_text= pad_sequences([token_text], maxlen=maxlen-1, padding='pre')

  # prediction
  position = np.argmax(model.predict(padded_token_text))

  for word, index in token.word_index.items():
    if index == position:
      text = text + ' ' + word
      print(text)
```
